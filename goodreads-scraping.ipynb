```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URLSTART = 'https://www.goodreads.com'
BESTBOOKS = '/list/show/21995.Best_21st_Century_Non_Fiction?page='
```

```python
url1 = URLSTART + BESTBOOKS +'1'
url1
```

```python
response1 = requests.get(url1)
response1.text[:500]
```

```python
soup1 = BeautifulSoup(response1.text, 'html.parser')
```

```python
soup1
```

```python
book_title = []
for book in soup1.find_all('a', class_='bookTitle'):
    title = book.text.strip()
    book_title.append(title)
    
book_title
```

```python
def get_title(book_node):
    return book_node.find('a', class_='bookTitle').text.strip()
```

```python
author_urls = []
for author in soup1.find_all('a', class_='authorName'):
    author_urls.append(author['href'])
    
author_urls
```

```python
def get_author_urls(authors_node):
    return authors_node.find('a', class_='authorName')['href']
```

```python
test = soup1.find('span', class_='minirating').text
test
```

```python
avg_rating = []
rating_count = []
for ratings in soup1.find_all('span', class_='minirating'):
    rating = ratings.text.strip().split(' ')
    avg_rating.append(rating[0])
    rating_count.append(rating[4])
```

```python
print(avg_rating, rating_count)
```

```python
def get_avg_rating(rating_node):
    return rating_node.find('span', class_='minirating').text.strip().split(' ')[0]
```

```python
def get_rating_count(rating_count_node):
    return rating_count_node.find('span', class_='minirating').text.strip().split(' ')[4]
```

```python
x = soup1.find('a', href='#').text.strip()
x
```

```python
score_list = []
for scores in soup1.find_all('span', class_='smallText uitext'):
    sc = scores.find('a', href='#').text.split(':')[1].strip()
    score_list.append(sc)

score_list
```

```python
def get_score(score_node):
    return score_node.find('a', href='#').text.split(':')[1].strip()
```

```python
books = []
for i in range(1, 6):
    URLSTART = 'https://www.goodreads.com'
    BESTBOOKS = '/list/show/21995.Best_21st_Century_Non_Fiction?page='
    url = URLSTART + BESTBOOKS + str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for node in soup.find_all('tr', itemtype='http://schema.org/Book'):
        book_info = {}
        book_info['Title'] = get_title(node)
        book_info['Author_url'] = get_author_urls(node)
        book_info['Avg_rating'] = get_avg_rating(node)
        book_info['Rating_count'] = get_rating_count(node)
        book_info['Score'] = get_score(node)
        books.append(book_info)
```

```python
df_books_raw = pd.DataFrame(books)
df_books_raw
```

```python
df_books_raw.to_csv('books_raw.csv', index=False)
```
